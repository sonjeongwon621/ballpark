### Summary  


##### The focus of this paper is twofold.
- Create a Brock Mirman type growth model with uninsured idiosyncratic income shocks and liquidity constraints.
- Quantitatively assess the relative importance of individual precautionary motives for aggregate saving.

##### Empirical evidence in favor of uninsured idiosyncratic risk (heterogeneity due to incomplete markets)
- Individual consumption is much more variable than aggregate consumption
- Individual consumption patterns are not highly correlated with each other or with aggregate consumption
- High mobility across wealth deciles
- Wide disparities in portfolio composition depending on wealth

##### Model Setup
Households solve the following

$$
\begin{aligned}
\max \quad &\mathbb{E}_0 [\Sigma^{\infty}_{t=0} \beta^t U(c_t)]\\
s.t. \quad &c_t + a_{t+1} = wl_t + (1+r)a_t;\\
&c_t \geq 0, a_t \geq -b \quad (a.s.)\\
&\lim_{t \to \infty} a_t/(1+r)^t \geq 0
\end{aligned}
$$

Uninsured idiosyncratic risk is introduced by allowing labor supply to be stochastic.

$$
l_t \text{ is i.i.d.}\\
l_t \in [l_{min}, l_{max}]\text{ with } l_{min} > 0
$$

##### Model Solution
By introducing new variables to represent the borrowing limit ($\Phi$), available asset ($\hat{a_t}$), and total resources ($z_t$) the household's problem can be represented by the following Bellman equation.

$$
\begin{aligned}
V(z_t,b,w,r) &\equiv \max_{\hat{a_{t+1}}} \{ U(z_t-\hat{a_{t+1}}) + \beta \int V(z_{t+1},b,w,r)dF(l_{t+1})\} \\
s.t. \quad z_{t+1} &= wl_{t+1} + (1+r)\hat{a_{t+1}}-r\Phi \\
\hat{a_t} &= a_t + \Phi \\
\Phi &= \left \{
  \begin{aligned}
    &\min \{ b, wl_{min}/r\} && \text{if}\ r>0 \\
    &b && \text{otherwise}
  \end{aligned} \right.
  \end{aligned}
$$

This results in the optimal asset choice of households.

$$
\hat{a_{t+1}}= A(z_t,b,w,r)
$$

By letting $f(k,1)$ denote per capita output as a function of per capita capital ($k$) and letting $\delta$ be the depreciation rate we can solve for the equilibrium. Note that $r$ is chosen endogenously.

##### Model Specifications

- $\beta = 0.96$
- $f(.)$ is assumed to be Cobb-Douglas with $\alpha=0.36$
- $\delta = 0.08$
- CRRA utility with $\mu \in \{1,3,5\}$ as the relative risk aversion coefficient
- $\log(l_t) = \rho \log(l_{t-1}) + \sigma(1-\rho^2)^{1/2}\epsilon_t$, $\epsilon_t \sim N(0,1)$, $\sigma \in \{0.2, 0.4\}$, $\rho \in \{0,0.3,0.6,0.9\}$
- $b=0$
